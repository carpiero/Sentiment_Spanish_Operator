```python
def hashtags(x):
    return x['hashtags']

def urls(x):
    if len(x['urls'])>0:
        return x['urls'][0]['url']
    else:
        return 'no url'

    
df_postgresql['hashtags']=df_postgresql['entities'].apply(hashtags)
df_postgresql['urls']=df_postgresql['entities'].apply(urls)






df_postgresql.info(memory_usage='deep')
```

```python
from transformers import pipeline

print(pipeline('sentiment-analysis')('we love you'))
```

```python
classifier = pipeline('sentiment-analysis', model='nlptown/bert-base-multilingual-uncased-sentiment')
```

```python
classifier('muy malo')
```

```python

list_tweet=list(df_postgresql['Tweet_Content'])

cl=list_tweet

len(cl)
```

```python

jump=2000
cla=[classifier(cl[x:x+jump]) for x in range(0,len(cl),jump)]
cla=[y for x in cla for y in x]

    
len(cla)
```

```python
def rem_stars(x):
    m=re.sub('stars|star','',x)
    return int(m)
```

```python
cl_label=[rem_stars(x.get('label')) for x in cla]
cl_score=[x.get('score') for x in cla]



len(cl_label)
```

```python
df_postgresql['stars'] = pd.Series((cl_label), index=df_postgresql.index)
df_postgresql['stars_score'] = pd.Series((cl_score), index=df_postgresql.index)
```
